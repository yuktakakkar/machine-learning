```python
#Plot atleast 20 visualization for the same dataset with different data points. Ensure that the type of visualization is appropriate and follows this chart .

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:\Users\Yukta\Desktop\data\weather.csv')
```
